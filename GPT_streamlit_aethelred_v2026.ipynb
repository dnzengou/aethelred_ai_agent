## The AI Assistant Build

### Aethelred, the Noble Advisor



```python
# aethelred_v3.py — run locally or in Colab with OpenAI / Grok / Anthropic API
import os
from openai import OpenAI  # or anthropic / groq / etc.

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))  # swap endpoint as needed

SYSTEM_PROMPT = """[Paste the full meta-prompt above here]"""

def aethelred(query: str, history: list = None) -> str:
    messages = [{"role": "system", "content": SYSTEM_PROMPT}]

    # Optional short memory
    if history:
        messages.extend(history[-4:])  # last 2 turns

    messages.append({"role": "user", "content": query})

    response = client.chat.completions.create(
        model="gpt-4o" or "grok-beta" or "claude-3.5-sonnet-20241022",
        messages=messages,
        temperature=0.65,          # balanced reasoning vs creativity
        max_tokens=1200
    )

    answer = response.choices[0].message.content

    # Optional: extract follow-up questions via simple regex or second call
    return answer

# Simple REPL loop (notebook / script)
print("Aethelred v3 — Noble Counsel. Describe your challenge.")
history = []

while True:
    user_input = input("\nYou: ")
    if user_input.lower() in ["exit", "quit", "bye"]:
        print("Farewell.")
        break

    reply = aethelred(user_input, history)
    print("\nAethelred:", reply)

    # Append to history
    history.append({"role": "user", "content": user_input})
    history.append({"role": "assistant", "content": reply})
```
